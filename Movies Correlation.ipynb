```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df_dataset = pd.read_csv(r'C:\Users\Abdulwahab Adeola\Documents\movies.csv')
```

```python
#Check for missing data
for col in df_dataset.columns:
    percent_missing = np.mean(df_dataset[col].isnull())
    print ('{} - {}%'. format(col, percent_missing))
```

```python
#Create release year column from the released date column
df_dataset['released_month'] = df_dataset['released'].astype(str).str[:4]
```

```python
df_dataset
```

```python
df_dataset.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Scatter plot budget vs gross

plt.scatter(x=df_dataset['budget'], y=df_dataset['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Budgets')
plt.show()
```

```python
#Budget vs Gross with Seaborn regplot
sns.regplot(x='budget', y='gross', data=df_dataset, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
#Correlation
df_dataset.corr()
```

```python
#Correlation Heatmap
correlation_matrix = df_dataset.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Assign numeric value to that does not have a numeric data type
df_num = df_dataset

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
        
df_num
```

```python
correlation_matrix = df_num.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Unstack the values for readability
corr_mat = df_num.corr()
corr_pairs = corr_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlations = sorted_pairs[(sorted_pairs) > 0.5]
high_correlations
```
